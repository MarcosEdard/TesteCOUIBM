```python
# Importar Bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import plotly.express as px
print("Setup Completo")
```

```python
# Importar Base de Dados
base = pd.read_csv("wine-quality-white-and-red.csv")
```

```python
# Quantidade de Linhas e Colunas na Base de Dados
base.shape
```

```python
# Primeira Visualização dos Dados (5 primeiras linhas)
base.head()
```

```python
base["type"].unique()
```

```python
# Checar valores NA
base.isnull().sum()
```

```python
# Informações Sobre os Dados(Quantidades de Valores e Tipos dos Dados)
base.info()
```

```python
# Mais informações
base.describe()
```

```python
# Boxplot Açúcar Residual
plt.figure(figsize = (10,5))
sns.set_theme(style = "whitegrid")
sns.boxplot(x = base["residual sugar"])
```

```python
# Boxplot Total Dioxido de Enxofre
plt.figure(figsize = (10,5))
sns.set_theme(style = "whitegrid")
sns.boxplot(x = base["total sulfur dioxide"])
```

```python
# Boxplot Livre de Dioxido de Enxofre
plt.figure(figsize = (10,5))
sns.set_theme(style = "whitegrid")
sns.boxplot(x = base["free sulfur dioxide"])
```

```python
# Colunas do DataSet
# Uso de "list" para legibilidade
list(base.columns)
```

```python
# Selecionar apenas dados numéricos para HeatMap
dados_heatmap = base.iloc[:,1:]
dados_heatmap
```

```python
# Gráfico (HeatMap)
plt.figure(figsize = (10,5))
sns.heatmap(dados_heatmap.corr(), annot = True)
```

```python
# Gráfico Teor Alcoólico X Qualidade (Scatterplot)
plt.figure(figsize = (10,5))
sns.scatterplot(x = "alcohol", y = "quality", hue = "type", data = base).set_title("Qualidade X Teor Alcoólico")
```

```python
# Gráfico Teor Alcoólico X Qualidade (barplot)
plt.figure(figsize = (10,5))
sns.barplot(x = "quality", y = "alcohol",hue = "type", data = base).set_title("Teor Alcoólico X Qualidade")
```

```python
# Gráfico Qualidade x Tipo (barplot)
plt.figure(figsize = (10,5))
sns.barplot(x = "type", y = "quality", data = base).set_title("Qualidade x Tipo")
```

```python
# Gráfico Densidade X Açúcar Residual (scatterplot)
plt.figure(figsize = (10,5))
sns.scatterplot(x = "density", y = "residual sugar",hue = "type", data = base).set_title("Densidade X Açúcar Residual")
```
